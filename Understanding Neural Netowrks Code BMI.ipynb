```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Generate artificial data
np.random.seed(42)  # For reproducibility

# Simulate 1000 data points for Age and BMI
X_1 = np.random.randint(20, 60, 1000)  # Age: between 20 and 60
X_2 = np.random.randint(18, 35, 1000)  # BMI: between 18 and 35

# Target label: Let's create a simple rule that decides if someone has diabetes
# We'll say if Age > 45 and BMI > 25, they are more likely to have diabetes (target = 1)
Y = ((X_1 > 45) & (X_2 > 25)).astype(int)  # Binary classification (0 or 1)

# Stack the features together to form the input dataset
X = np.column_stack((X_1, X_2))

# Visualize the data
plt.scatter(X_1, X_2, c=Y, cmap='bwr', edgecolor='k')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Artificial Data: Age vs BMI (Red=Diabetes, Blue=No Diabetes)')
plt.show()
```

```python
# Import TensorFlow libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Build the neural network model
model = Sequential()

# Input layer and hidden layer
model.add(Dense(4, input_dim=2, activation='relu'))  # 2 input features (Age and BMI), 4 neurons in hidden layer

# Output layer
model.add(Dense(1, activation='sigmoid'))  # 1 output neuron for binary classification

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on the artificial data
history = model.fit(X_train, Y_train, epochs=50, batch_size=10, validation_data=(X_test, Y_test))
```

```python
# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

```
